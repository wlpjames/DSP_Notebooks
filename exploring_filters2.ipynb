```python
import numpy as np
import matplotlib.pyplot as plt
```

we can take the ifft of a window function to get useable convolutionary data

```python
#high pass filter at 1000 hz
freq = 6000
samples = 48000
filt_freqDom = np.append(np.append(np.full(freq, 1), np.zeros(samples-(freq*2))), np.full(freq, 1))
filt_timeDom = np.fft.ifft(filt_freqDom)
plt.plot(filt_freqDom)
plt.show()
plt.plot(np.real(filt_timeDom))
plt.show()
```

im not realy sure of what to make of this, but it is theoretical noise in the time domain of <1000hz

### The suggested algorythm is,
$Y(z)=H(z)∗X(z)$

where Y is a Z-transform of the output, H is the Z-transform of the filter and X is the z-transform of the input

```python
#creating input
inpt = np.random.rand(samples) * 2 - 1

#apparently arctanh == fisher z transform
inptX = np.arctanh(inpt)

plt.plot(inpt)
plt.show()

plt.plot(inptX)
plt.show()
```

```python
#

outputY =  np.convolve(filt_timeDom, inptX)
plt.plot(outputY)


plt.show()
output = np.tanh(outputY)


plt.plot(output)
plt.show()
```

this seems a little dodge so i'll check it in freq domain

```python
output_in_freq = np.fft.fft(output)

plt.plot(inpt)
plt.show()

plt.plot(output_in_freq)
plt.show()
```

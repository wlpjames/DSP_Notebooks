```python
import numpy as np
import matplotlib.pyplot as plt

def han(n):
    
    out = np.zeros(n)
    for i in range(n):
        #do calc
        out[i] = .5 * (1 - np.cos(np.pi * 2 * (i / n)))
        
    return out
```

```python
N = 100

other = np.hanning(N)
mine = han(N)

for i in range(N):
    if other[i] != mine[i]:
        print ('unequal at: {}'.format(i))
        break

plt.plot(other)
plt.show()
plt.plot(mine)
plt.show()
```

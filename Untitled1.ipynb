### Example of FIR Filtering

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import wavio

```

### Here is some noise for looking at a freq responce

```python
#some noise:
sample_size = 44100
noise = np.random.rand(sample_size) - .5
plt.plot(noise)
plt.show()
plt.plot(np.fft.ifft(noise))
plt.show()
Audio(noise, rate=sample_size)
```

### Here is a wav file to use to modify


```python
from scipy.io.wavfile import read
cello = read('cello.wav')[1] / 5000
sample_rate = 44100
plt.plot(cello)
plt.show()
plt.plot(np.fft.ifft(cello))
plt.show()        
Audio(cello, rate=sample_rate)
```

```python

class FIR:
    def __init__(self, B, sampleRate, passes=1):
        self.B = B
        self.sample_rate = sample_rate
        self.passes = passes
        self.last = np.full(len(B), 0)
        
    def process(self, signal):
        B_len = len(self.B)
        signal_len = len(signal)
        
        #a temp save for the next last
        B = signal[:len(self.B)]
        
        #arrays for computing
        output = signal
        
        signal = np.append(self.last, signal)
        
        count = 0
        #iterating the signal
        for N in range(signal_len):
            count+=1
            sample = 0
            #iterating Beta values
            for K in range(B_len):
                sample += self.B[K] * signal[(N + B_len)-K]
                #print('n = {}; sample = {} val = {}'.format(N, sample, self.B[K] * signal[N-K]))
                #print(self.B[K] * signal[N-K])
            output[N] = sample
        
        print('self.B: {}'.format(self.B))
        print('signal: {}'.format(signal))
        print('{} samples processed'.format(count))
        print('output: {}'.format(output))
        
        return output
    
```

```python
#create lowpass window w. cuttoff at 4000
slope_len = 2000
cut_off = 1000
window = (np.concatenate([np.full(cut_off - (slope_len // 2), 1), 
                          np.hanning(slope_len * 2)[slope_len:], 
                          np.full(sample_rate - (slope_len // 2) - cut_off, 0)]))
plt.plot(window)
plt.show()
print(len(window))
```

```python
#the impulse responce
impulse_responce = np.fft.ifft(window)
plt.plot(impulse_responce[:300])
plt.show()
M = 40
B = np.real(impulse_responce[:M])
```

```python
#test this on noise
filt = FIR(B, sample_rate)
filtered_noise = filt.process(noise)
plt.plot(filtered_noise[:500])
plt.show()
plt.plot(np.fft.ifft(filtered_noise))
print(filtered_noise)
Audio(filtered_noise, rate=sample_rate)
```

```python
#test this on cello
filt = FIR(B, sample_rate)
filtered_cello = filt.process(cello)
plt.plot(filtered_noise[:500])
plt.show()
print(filtered_cello)
plt.plot(np.fft.ifft(filtered_cello))
plt.show()

Audio(filtered_cello, rate=sample_rate)
```

```python
#create highpass window w. cuttoff at 4000
slope_len = 2000
cut_off = 3000
hp_window = (np.concatenate([np.full(cut_off - (slope_len // 2), 0), 
                          np.hanning(slope_len * 2)[:slope_len], 
                          np.full((sample_rate // 2) - (slope_len // 2) - cut_off, 1)]))
hp_window = np.append(hp_window, np.flip(hp_window))
plt.plot(hp_window)
plt.show()
print(len(hp_window))
```

```python
#the impulse responce
impulse_responce = np.fft.ifft(hp_window)
plt.plot(impulse_responce[:300])
plt.show()
M = 40
B = np.real(impulse_responce[:M])
```

```python
#test this on noise
filt = FIR(B, sample_rate)
filtered_noise = - filt.process(noise)
plt.plot(filtered_noise[:500])
plt.show()
plt.plot(np.fft.ifft(filtered_noise))
print(filtered_noise)
Audio(filtered_noise, rate=sample_rate)
```

```python
#test this on cello
filt = FIR(B, sample_rate)
filtered_cello = filt.process(cello)



plt.plot(filtered_cello)
plt.show()

print('original')
plt.plot(np.fft.ifft(cello))
plt.show()

print('filtered')
plt.plot(np.fft.ifft(filtered_cello))
plt.show()

Audio(filtered_cello, rate=sample_rate)
```
